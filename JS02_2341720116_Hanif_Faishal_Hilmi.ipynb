**Langkah 1**

Memuat Data

```python
df = pd.read_csv('Titanic-Dataset.csv')

df.head()
```

**Langkah 2**

Inspeksi data singkat

```python
df.info()
df.isnull().sum()
```

**Langkah 3**

Data Imputation

Melakukan data imputation ke data yang hilang, yaitu "Age", "Cabin", dan "Embarked".

Cara yang digunakan :
* "Age" menggunakan strategi mean karena merupakan data nominal,
* "Cabin" menggunakan substitusi data dengan informasi "DECK" dengan asumsi seluruh penumpang tidak memilki nomor kabin,
* "Embarked" menggunakan modus karena hanya 2 data saja yang hilang

```python
# Age - mean
df['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin - "Deck"
df['Cabin'].fillna(value='DECK', inplace=True)

# Embarked - modus
df['Embarked'].fillna(value=df['Embarked'].mode, inplace=True)
```

**Langkah 4**

Validasi hasil

```python
df.head()
df.info()
```

**Langkah 2**

Resize Gambar

Mengubah ukuran gambar menjadi 128x128px

```python
img_resized = cv2.resize(img_rgb, (128, 128))

plt.imshow(img_resized)
plt.title('Gambar setelah Resized (128x128)')
plt.axis("off")
plt.show()
```

**Langkah 3**

Ubah ke Grayscale

Mengubah warna citra ke bentuk grayscale

```python
img_gray = cv2.cvtColor(img_resized, cv2.COLOR_RGB2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.title('Gambar Grayscale')
plt.axis("off")
plt.show()
```

**Jawab**

```python
import pandas as pd
```

**No. 1**

```python
df = pd.read_csv('wbc.csv')
df.info()
df.head()
```

Dari dataset tersebut, variabel "id" dan "Unnamed: 32" adalah variabel yang tidak dapat digunakan. Oleh karena itu, variabel tersebut dipisah dari dataframe.

```python
# hapus kolom Unnamed: 32
df = df.drop(columns=['id', 'Unnamed: 32'])
df.head()
```

**No. 2**

Melakukan encoding pada kolom "diagnosis"

```python
from sklearn.preprocessing import LabelEncoder

# insialisasi LabelEncoder
le = LabelEncoder()

# Encoding ke kolom diagnosis
df['diagnosis'] = le.fit_transform(df['diagnosis'])

df.head()
```

**No. 3**

Melakukan standardisasi pada semua kolom yang memiliki nilai numerik

```python
from sklearn.preprocessing import StandardScaler

# memilih kolom numerik selain diagnosis dan id
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
numeric_cols = [col for col in numeric_cols if col not in ['diagnosis']]

# inisialisasi StandardScaler
scaler = StandardScaler()

# standardisasi ke kolom numerik
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

df.head()
```
